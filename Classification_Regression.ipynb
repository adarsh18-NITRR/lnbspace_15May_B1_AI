```python
df2 = df[:100]
```

```python
df2
```

```python
sns.pairplot(df2, hue='species')
#now here we only two classes and hence the data is binary
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain ,xtest, ytrain, ytest = train_test_split(df2.drop('species', axis=1), df2['species'], test_size=0.2, random_state=50)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
#we get 100 beacuse the data was quiet seperated from each other
```
