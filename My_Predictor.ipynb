```python
import pandas as pd
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('used_cars_data.csv', index_col=0)
```

```python
df.info()
```

```python
import numpy as np
```

```python
# we use pd.null beause it gives truth value for the dataframe
#however the np.nan returns null value

def converter(x):
    if not pd.isnull(x):
        try:
            return float(x.split()[0])
        except Exception:
            return np.nan
    else:
        return np.nan
```

```python
df['Power'] = df['Power'].apply(converter)
```

```python
#df['Price'] = df['Price'].apply(converter)
```

```python
df['Engine'] = df['Engine'].apply(converter)
```

```python
df['Mileage'] = df['Mileage'].apply(converter)
```

```python
df['Name'] = df['Name'].apply(lambda x:x.split()[0])
```

```python
df['Name'].str.get_dummies()
```

```python

df['Owner_Type'].str.get_dummies()
df['Fuel_Type'].str.get_dummies()
df['Transmission'].str.get_dummies()
```

#### <b>Here we are going to use get_dummies to make string columns with categorical data useful for ML

```python
df['Manual'] = pd.get_dummies(df['Transmission'], drop_first=True)
```

```python
del df['Transmission']
```

```python
df = pd.concat([df,df['Fuel_Type'].str.get_dummies()], axis=1)
```

```python
del df['Fuel_Type']
```

```python
df = pd.concat([df,df['Owner_Type'].str.get_dummies()], axis=1)
```

```python
del df['Owner_Type']
```

#### <b>Here we are going to take name, location and year into consideration

```python
df = pd.concat([df,df['Name'].str.get_dummies()], axis=1)
```

```python
del df['Name']
```

##### <b>Making convertor for converting year float values to string values

```python
df['Year'] = df['Year'].apply(lambda x:str(x))
```

```python
df = pd.concat([df,df['Year'].str.get_dummies()], axis=1)
```

```python
del df['Year']
```

```python
df = pd.concat([df,df['Location'].str.get_dummies()], axis=1)
```

```python
del df['Location']
```

```python
df.columns
```

### <b>here we are deleting new_price

```python
df.dropna(axis=0, inplace=True)
```

#### <b>Now we are going to feed the data into the model

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.5, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
