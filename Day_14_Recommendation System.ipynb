#### <b>unique, mean, groupby, sort_values</b>

```python
data3['rating'].mean() #this is the complete average rating
```

```python
data3['movie_name'].nunique()
    
```

```python
for i in data3['movie_name'].unique():
     print(i, data3['rating'][data3['movie_name']==i].mean())
#In the for loop we have name of all the movies by using unique functions
#In print statement we have name of movies and rating mean if the movie_name is == i( whichi is movie name only)
# and then calculate the mean of the rating
```

```python
# same above thing using groupby
# here we group the data by movie_name - then find there rating mean - the sort ratings in asending order
# head gave us the top twenty 
data3.groupby('movie_name')['rating'].mean().sort_values(ascending=False).head(20)
```

```python
## This is creating a table from data3 using pivot_table functions - 
##### The result table is going to have
###index as data3's user_id
###column names as data3's movie_names
###and value in the tables as 
data_pivot = data3.pivot_table(index='user_id', columns='movie_name', values='rating')
```

```python
data_pivot
```

```python
movie_r.sort_values('rating',ascending=False).head(20)
```

```python
# now here we are going to fetch what are the ratings given to the movie1 by each user
```

```python
#criminal_movie_mat = data_pivot['They Made Me a Criminal (1939)']

#This is the table in which all the user's rating for the specific movie are given
s_list_mat = data_pivot['Schindler\'s List (1993)']
```

```python
s_list_mat.head(10)
```

#### <b>Using'corrwith' function

```python
# now we are going to find out what are the movies that have similar ratings like this
# i.e. relating the crimianl_movie_mat with the entire data
```

```python
s_list_similar = pd.DataFrame(data_pivot.corrwith(s_list_mat),columns=['correlation'])
```

```python
s_list_similar.dropna(inplace=True)
```

```python
s_list_similar.sort_values(by='correlation',ascending=False)
```

```python
#now we are going to club similarity/correlation with number of rating that the movie has got
```

```python
s_list_similar['count'] = data3['movie_name'].value_counts()
```

```python
s_list_similar
```

```python
s_list_similar[s_list_similar['count']>100].sort_values(by='correlation',ascending=False).head(3)
```

### <b>All of the recommendation code together

```python
user_movie_input = input('Enter a movie name for which you want recommendation')
s_list_mat = data_pivot[user_movie_input]
s_list_similar = pd.DataFrame(data_pivot.corrwith(s_list_mat),columns=['correlation'])
s_list_similar.dropna(inplace=True)
s_list_similar.sort_values(by='correlation',ascending=False)
s_list_similar['count'] = data3['movie_name'].value_counts()
s_list_similar[s_list_similar['count']>100].sort_values(by='correlation',ascending=False).head(3)
```
