```python
import pandas as pd 
import numpy as np
```

```python
from keras.datasets import boston_housing as bh
```

```python
(xtrain , ytrain), (xtest, ytest) = bh.load_data()
```

```python
featrues = '''crim
zn
indus
chas
nox
rm
age
dis
rad
tax
ptratio
black
lstat
medv'''
```

```python
xtrain[2] 
#- the values for individaul vales
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(activation='relu',input_shape=(13,), units=48))
model.add(Dense(activation='relu', units=65))
model.add(Dense(activation='relu', units=65))
model.add(Dense(activation='relu', units=65))
model.add(Dense(activation='relu', units=65))
model.add(Dense(activation = 'relu', units = 1))
model.compile(optimizer='rmsprop', loss='mse', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain ,epochs=20)
```

```python
from keras.callbacks import EarlyStopping
#to stop after the best epoch - least loss
```

```python
es = EarlyStopping(monitor = 'mse',patience=20)
#monitoring - validatoin loss
# pateince = 0 - if not imporve in the next epoch stop
```

```python
h = model.fit(xtrain, ytrain ,epochs=1000,callbacks=[es])
```

```python
model.save('bhmodel')
#can save keras model
```
