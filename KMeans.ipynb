```python
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sklearn.datasets as skd
```

```python
data = skd.make_blobs(n_samples=500, n_features=2, centers=3, cluster_std=1.2, random_state=101)
#2 features
#four centers - will be there for grouping
#cluster_std = 1 - not scattere data will be there

#the end array will be there for labels - 0,1,2,3 - as there were 4 centres
```

```python
#we are not going to use the labels
```

```python
df = pd.DataFrame(data[0])
y = pd.Series(data[1])
df['y'] = y
```

```python
df.groupby(by='y').describe().T
```

```python
sns.scatterplot(df[0],df[1], c=df['y'])
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3).fit(df.drop('y',axis=1))
```

```python
#here we have not given the label y column
```

```python
sns.scatterplot(df[0],df[1], c=model.labels_)
```

```python
###Here KMeans has seperated the data in the clusters as they were
```
